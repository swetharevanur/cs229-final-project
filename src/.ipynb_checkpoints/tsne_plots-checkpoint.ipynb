```python
import pandas as pd
import os
from yellowbrick.text import TSNEVisualizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

```python
# read in cleaned corpus
path_to_textf = os.path.join('../cleaned_data', 'cleaned_text_tsne.pkl')
corpus = pd.read_pickle(path_to_textf)
```

```python
# load tf-idf vectorizer
tfidf = TfidfVectorizer()
docs = tfidf.fit_transform(corpus['text'].values)
labels = corpus['class'].values
```

```python
tsne = TSNEVisualizer()
tsne.fit(docs, labels)
tsne.poof()
```

```python
# apply clustering instead of class names
clusters = KMeans(n_clusters = 6)
clusters.fit(docs)

tsne = TSNEVisualizer()
tsne.fit(docs, ["c{}".format(c) for c in clusters.labels_])
tsne.poof()
```
